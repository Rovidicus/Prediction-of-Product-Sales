```python
# imports
import pandas as pd
pd.set_option('display.max_columns',100)
# Saving the filepath and assigning dataframe as df
fpath = 'Data/sales_predictions_2023.csv'
df = pd.read_csv(fpath)
df.head()
```

```python
df=df.drop(columns = ["Item_Identifier","Outlet_Identifier","Outlet_Establishment_Year"], axis=1)
df.head(3)
```

```python
# Correcting inconsistencies in Item_Fat_Content
df['Item_Fat_Content'].value_counts()
```

```python
df["Item_Fat_Content"] = df['Item_Fat_Content'].str.replace('LF', 'Low Fat').str.replace('low fat', 'Low Fat').str.replace('reg', 'Regular')
df['Item_Fat_Content'].value_counts()
```

```python
# Saving dataset
df.to_csv("Data/sales_prepped.csv", index=False)
```
